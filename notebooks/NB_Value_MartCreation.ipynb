# NB_Value_MartCreation
Final aggregation for reporting layer (Gold).

```python
from pyspark.sql.functions import current_timestamp

# Sample Value logic: Final Gold table for Power BI
try:
    # 1. Read from Entity Table
    df_entity = spark.table("claims_entity")
    
    # 2. Add Mart creation metadata
    df_gold = df_entity.withColumn("gold_timestamp", current_timestamp())
    
    # 3. Save to Gold Table
    df_gold.write.format("delta").mode("overwrite").saveAsTable("claims_gold")
    print("Success: claims_gold updated for Power BI.")
    
except Exception as e:
    print(f"Gold processing failed: {str(e)}")
```
