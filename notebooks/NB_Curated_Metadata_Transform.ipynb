# NB_Curated_Metadata_Transform
Apply standard curation (de-duplication, basic cleaning) to bronze tables.

```python
from pyspark.sql.functions import current_timestamp, upper

# Sample Curation logic for LH_Claims
try:
    # 1. Read from Bronze Table
    df_bronze = spark.table("claims_bronze")
    
    # 2. Basic Transformations (Upper Case source field, remove duplicates)
    df_silver = df_bronze.dropDuplicates()\
                         .withColumn("curation_timestamp", current_timestamp())\
                         .withColumn("domain", upper(lit("CLAIMS")))
    
    # 3. Save to Silver Table
    df_silver.write.format("delta").mode("overwrite").saveAsTable("claims_silver")
    print("Success: claims_silver updated.")
    
except Exception as e:
    print(f"Curation failed: {str(e)}")
```
